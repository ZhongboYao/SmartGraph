```python
import os
import sys
sys.path.append(os.path.join(os.path.abspath(""), "..", "backend"))

from pathlib import Path
from pdf_reading import content_extraction
from graph_extraction import extract_graph, to_cytoscape
from graph_store import GraphStore
from merge import merge_by_label
```

```python
pdf1 = Path("../data/notes/GPT.pdf")
pdf2 = Path("../data/notes/BPE.pdf")

content_save_path = Path("../log/extracted_notes")
claude_graph_save_path = Path("../log/graphs_from_claude")
cyto_graph_save_path = Path("../log/graphs_for_cyto")
final_graph_path = Path("../data/graphs/graph.json")

def note_update_pipeline(pdf_path: Path):
    content = content_extraction(pdf_path, True, content_save_path)
    graph = extract_graph(content, pdf1.stem, True, claude_graph_save_path)
    cyto_dict = to_cytoscape(graph, True, cyto_graph_save_path)

    graphstore = GraphStore()
    graphstore.load(final_graph_path)
    graphstore.merge(cyto_dict, merge_by_label)
    graphstore.save(final_graph_path)

note_update_pipeline(pdf1)
note_update_pipeline(pdf2)
```
